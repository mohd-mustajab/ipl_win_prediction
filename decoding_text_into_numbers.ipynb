```python
import pandas as pd

# Load IPL match data
df = pd.read_csv("../superhero_data.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in ['Hero_Name', 'Real_Name', 'Power', 'Origin', 'Alignment', 'Base_City']:
    df[col] = le.fit_transform(df[col])

df.head()
```
